# MapLibreum: Basic Usage

This notebook covers the basic usage of MapLibreum, including map initialization, setting the initial view, adding simple markers, popups, and basic geometry.

## 1. Initializing a Map

The most basic MapLibreum map can be created by simply instantiating the `Map` class. By default, it will center on [0, 0] with a low zoom level.

```
from maplibreum import Map

# Create a basic map
m1 = Map()

# Display the map
m1
```

## 2. Setting the Initial View

You can specify the initial center coordinates `[longitude, latitude]` and the zoom level when creating the map.

```
# Create a map centered on Paris, France
m2 = Map(center=[2.3522, 48.8566], zoom=12)

m2
```

## 3. Adding Markers

Markers are used to identify locations on the map. You can add default markers with the `Marker` class.

```
from maplibreum import Marker, Tooltip

# Map centered on New York City
m3 = Map(center=[-74.0060, 40.7128], zoom=11)

# Add a simple marker
Marker(coordinates=[-74.0060, 40.7128]).add_to(m3)

# Add another marker with a tooltip (text shown on hover)
Marker(
    coordinates=[-73.9855, 40.7580], 
    tooltip=Tooltip("Times Square")
).add_to(m3)

m3
```

## 4. Popups

Popups are info windows that open when a user clicks on a marker. They can contain simple text or HTML content.

```
from maplibreum import Popup

m4 = Map(center=[-118.2437, 34.0522], zoom=10) # Los Angeles

# Marker with simple string popup
Marker(
    coordinates=[-118.2437, 34.0522], 
    popup=Popup(html="Downtown LA")
).add_to(m4)

# Marker with HTML popup and specific dimensions
html_content = """
    <h4>Hollywood Sign</h4>
    <p>Famous landmark in the Hollywood Hills.</p>
"""
popup = Popup(html=html_content)

Marker(
    coordinates=[-118.3215, 34.1341], 
    popup=popup
).add_to(m4)

m4
```

## 5. Basic Geometry (Vectors)

You can draw basic shapes on the map using vector layers: `PolyLine`, `Polygon`, `Circle`, and `CircleMarker`.

```
from maplibreum.core import PolyLine, Polygon, Circle, CircleMarker

m5 = Map(center=[-0.1278, 51.5074], zoom=12) # London

# 1. PolyLine (a line connecting multiple points)
line_coords = [
    [-0.1419, 51.5014], # Buckingham Palace
    [-0.0759, 51.5081], # Tower of London
    [-0.1195, 51.5033]  # London Eye
]
PolyLine(
    locations=line_coords,
    color="blue",
    weight=5,
    tooltip=Tooltip("Sightseeing Route")
).add_to(m5)

# 2. Polygon (a filled shape)
polygon_coords = [
    [[-0.1764, 51.5098], # Hyde Park points
    [-0.1541, 51.5065],
    [-0.1834, 51.5042],
    [-0.1764, 51.5098]]  # Close the polygon (Note nested list for outer ring)
]
Polygon(
    locations=polygon_coords,
    color="green",
    fill=True,
    fill_color="lightgreen",
    fill_opacity=0.5,
    tooltip=Tooltip("Hyde Park Area")
).add_to(m5)

# 3. Circle (radius is in meters)
Circle(
    location=[-0.0759, 51.5081], # Tower of London
    radius=500, # 500 meters
    color="red",
    fill=True,
    fill_color="pink"
).add_to(m5)

# 4. CircleMarker (radius is in pixels, stays same size when zooming)
CircleMarker(
    location=[-0.1195, 51.5033], # London Eye
    radius=10, # 10 pixels
    color="purple",
    fill=True,
    fill_color="purple"
).add_to(m5)

m5
```
